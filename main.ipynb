```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

trips=np.array(pd.read_csv(r"/home/aziza/Downloads/coding/friendly-funicular/trips.csv",encoding='cp866'))
print(trips)
```

```python
T=np.zeros([22,22])
O=np.zeros([22])
D=np.zeros([22])
z=0
for i in range(22):
    O[i]=Counter(trips.T[0])[i+1]
    D[i]=Counter(trips.T[1])[i+1]
    tr=np.zeros([22])
    print(z,int(O[i])+z-1)
    for j in range(z,int(O[i])+z):
        tr[int(trips[j][1])-1]=trips[j][2]
        if j==int(O[i])+z-1:
            z=int(O[i])+z
    print(tr)
    T[i]=tr
print(T)
```
